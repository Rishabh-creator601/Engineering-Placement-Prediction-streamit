```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier 
import pickle
```

```python
data = pd.read_csv("collegePlace.csv")
```

```python
data.head()
```

```python
data["Stream"].value_counts()
```

```python
streams = []
genders  = []
```

```python
le = LabelEncoder()
data["Stream_up"] =le.fit_transform( data["Stream"])
streams.extend(le.classes_)
data["Gender_up"] =le.fit_transform( data["Gender"])
genders.extend(le.classes_)
```

```python
genders
```

```python
data.head()
```

```python
x = data.drop(columns=["Gender","Stream","PlacedOrNot"],axis=1)
y = data["PlacedOrNot"]
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_train
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
predictions = model.predict(x_test)
print(classification_report(y_test,predictions))
```

```python
with open("model.pkl","wb") as f:
    pickle.dump(model,f)

print("Model Saved")
```
